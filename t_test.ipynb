4. Xây dựng một kiểm định t cụ thể với số liệu được sinh ngẫu nhiên theo phân phối chuẩn
Bài toán : Ta có giả thiết, chiều cao trung bình của 100 chiều cao trung bình của người trưởng thành từ 20 tuổi trở lên là khoảng 66,5 inch (69,3 inch đối với nam) với độ lệch chuẩn là 5,3. Ta sẽ kiểm tra xem khẳng định này có phù hợp với mức ý nghĩa là 5% hay không? 
**Kiểm định 1 đuôi**

Xây dựng giả thuyết:

$$H_0 : \mu = 66,5$$

Giả thuyết thay thế:

$$H_a : \mu < 66,5$$

với:
- Hệ số alpha : 0,05  
- Chỉ số t tương ứng với hệ số alpha/2 là : 1.984

Công thức:  
$$t = \frac{(\overline{\rm X} - \mu_0)}{(s / \sqrt{n})}$$
với:
- $\overline{\rm X}$: trung bình mẫu 
- $\mu_{0}$: trung bình của tổng thể cần so sánh 
- s : độ lệch chuẩn hiệu chỉnh mẫu 
- n : cỡ mẫu

Áp dụng với các giá trị nêu trên:
- $\overline{\rm X}$: 66,5 
- $\mu_{0}$: 69,3
- s : 5,3
- n: 100
$$t = \frac{(66,5- 69,3)}{(5.3 / \sqrt{100})} \approx -5,28$$

Giá trị $|t|$ = -5,28 < 1,984 nên không có bằng chứng để bác bỏ giả thuyết $H_0$


Lập trình python cho bài toán trên


```python
#Sinh ngẫu nhiên một mẫu 100 phần tử có trung bình mẫu là 66,5, độ lệch chuẩn là 5,3 
samples = np.random.normal(66.5, scale =5.3,size = 100)
print(samples[:10])
```

```python
#Xác định các thông số
null_mean = 100
alpha = 0.05
```

```python
#Kiem dinh t 
t_test, p_value= stats.ttest_1samp(samples, null_mean)
print("p_value: ",p_value)
if(p_value < alpha):
  print("Có bằng chứng để bác bỏ giả thuyết H0")
else:
  print("Chưa có bằng chứng để bác bỏ giả thuyết H0")
```

**Kiểm định giả thuyết 2-đuôi**
Bài toán : Ta có giả thiết, chiều cao trung bình của 100 chiều cao trung bình của người trưởng thành từ 20 tuổi trở lên là khoảng 66,5 inch (69,3 inch đối với nam với độ lệch chuẩn là 5,3 và 63,8 inch đối với nữ với độ lệch chuẩn là 4.8. Ta sẽ kiểm tra xem khẳng định này có phù hợp với mức ý nghĩa là 5% hay không? 
Xây dựng giả thuyết:
\begin{align}
        H_0: μ_1 = μ_2
    \end{align}

Giả thuyết thay thế:
\begin{align}
        H_a: μ_1 \ne μ_2
    \end{align}

với:
- Hệ số alpha : 0,05  
- Chỉ số t tương ứng với hệ số alpha/2 là : 1.984




Lập trình python cho bài tập trên 

```python
#Sinh ngẫu nhiên sample 1 có 50 phần tử với mean = 69,3; std = 5,3 
samples_1 = np.random.normal(69.3,scale= 5.3, size = 50)
print(samples_1[:10])
```

```python
#Sinh ngẫu nhiên sample 2 có 50 phần tử với mean = 63,8;  std = 4,8 
samples_2 = np.random.normal(63.8, scale=4.8,size= 50)
print(samples_2[:10])
```

```python
#Kiểm định giả thuyết 
alpha = 0.05
t_test, p_value,df = ttest_ind(x1 = samples_1, x2 = samples_2, value = 69.3-63.8 , alternative = 'two-sided')
print('t = ', t_test, '; p = ', p_value)
if(p_value <  alpha):
  print("Có bằng chứng để bác bỏ giả thuyết H0")
else:
  print("Chưa có bằng chứng để bác bỏ giả thuyết H0")
```
