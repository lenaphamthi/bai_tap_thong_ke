Xây dựng đoạn code lập trình permutation test


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import random as random
```

Đọc dữ liệu


```python
session_times = pd.read_csv("https://raw.githubusercontent.com/gedeck/practical-statistics-for-data-scientists/master/data/web_page_data.csv")
```

Lệnh boxplot pandas sử dụng đối số từ khóa by để tạo hình:

```python
ax = session_times.boxplot(by='Page', column='Time')
ax.set_xlabel('')
ax.set_ylabel('Time (in seconds)')
plt.suptitle('')
```

lọc khung dữ liệu cấu trúc theo từng trang và sau đó xác định giá trị trung bình
của cột Thời gian:

```python
#Trang A 
mean_a = session_times[session_times.Page == 'Page A'].Time.mean()
print(mean_a)
```

```python
#Trang B
mean_b = session_times[session_times.Page == 'Page B'].Time.mean()
print(mean_b)
```

```python
mean_b - mean_a
```

Permutation test 

```python
def perm_fun(x = mean_a, nA=21, nB=15):
  n = nA + nB
  idx_B = set(random.sample(range(n), nB))
  idx_A = set(range(n)) - idx_B
  return x.loc[idx_B].mean() - x.loc[idx_A].mean()
perm_fun
```

```python
perm_diffs = [perm_fun(session_times.Time, nA=21, nB=15 ) for _ in range(1000)]
fig, ax = plt.subplots(figsize=(5, 5))
ax.hist(perm_diffs, bins=11, rwidth=0.9)
ax.axvline(x = mean_b - mean_a, color='black', lw=2)
ax.text(50, 190, 'Observed\ndifference', bbox={'facecolor':'white'})
ax.set_xlabel('Session time differences (in seconds)')
ax.set_ylabel('Frequency')
```
